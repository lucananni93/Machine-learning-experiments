## Dimensionality curse

```python
def edge_length(fraction_of_observations, n_dimensions):
    return np.power(fraction_of_observations, 1/n_dimensions)
```

```python
plt.figure(figsize=(7, 7))
for p in np.arange(1, 11, dtype=int):
    pp = []
    for fo in np.arange(0.01, 1.0, 0.05):
        el = edge_length(fo, p)
        pp.append(el)
    
    plt.plot(np.arange(0.01, 1.0, 0.05), pp, label="p = {}".format(p))
plt.legend()
plt.xlabel("Fraction of volume")
plt.ylabel("Edge")
plt.show()
```
