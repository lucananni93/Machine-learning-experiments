```python
import seaborn as sns
import matplotlib.pyplot as plt

import numpy as np
import scipy.stats as ss

%matplotlib inline
```

# Binary Variables

```python
n_samples = 1000
```

## Bernoulli
$$
    \text{Bern}(x | \mu) = \mu^x\left( 1 - \mu \right)^{1 - x}
$$

```python
p = 0.6

samples = np.random.binomial(1, p, n_samples)
sns.distplot(samples)
mean = np.mean(samples)
std = np.std(samples)
print("Mean: {}\nStd: {}".format(mean, std))
```

## Binomial
$$
    \text{Bin}(m | N, \mu) = \left( \begin{array}{c} N\\m \end{array} \right)\mu^m\left( 1 - \mu \right)^{m - N}
$$

```python
p = 0.6
m = 5

samples = np.random.binomial(m, p, n_samples)
sns.distplot(samples)
mean = np.mean(samples)
std = np.std(samples)
print("Mean: {}\nStd: {}".format(mean, std))
```

## Beta
$$
    \text{Beta}(\mu|a, b) = \frac{\Gamma(a + b)}{\Gamma(a)\Gamma(b)}\mu^{a - 1}\left( 1 - \mu \right)^{b - 1}
$$
$$
    \Gamma(x) = \int_0^\infty u^{x - 1}e^{-u} du
$$

```python
a = 2
b = 2

samples = np.random.beta(a, b, n_samples)
sns.distplot(samples)

mean = np.mean(samples)
std = np.std(samples)
print("Mean: {}\nStd: {}".format(mean, std))
```

# Multinomial variables

## Dirichlet
$$
    \text{Dir}(\boldsymbol{\mu} | \boldsymbol{\alpha}) = \frac{\Gamma(\alpha_0)}{\Gamma(\alpha_1)\cdots\Gamma(\alpha_K)}\prod_{k=1}^K \mu^{\alpha_k - 1}_k, \qquad \alpha_0 = \sum_{k = 1}^K \alpha_k
$$

```python
alpha = [2, 2]

samples = np.random.dirichlet(alpha, n_samples)
sns.jointplot(x=samples[:, 0], y=samples[:, 1], kind="kde")
```

## Gaussian
$$
    \mathcal{N}(\mathbf{x} | \boldsymbol\mu, \boldsymbol\Sigma) = \frac{1}{(2\pi)^{\frac{D}{2}}}\frac{1}{|\boldsymbol\Sigma|^{\frac{1}{2}}}\text{exp}\left\{ -\frac{1}{2}(\mathbf{x} - \boldsymbol\mu)^T \boldsymbol\Sigma^{-1}(\mathbf{x} - \boldsymbol\mu) \right\}, \qquad \boldsymbol\Sigma = D\times D \text{ covariance matrix}
$$

```python
mu = np.array([0.1, 0.5])
t = np.random.rand(2,2)
sigma = t.dot(t.T)

samples = np.random.multivariate_normal(mu, sigma, n_samples)
sns.jointplot(x=samples[:, 0], y=samples[:, 1], kind="kde")
```

## Gaussian Mixture models

    p(\mathbf{x}) = \sum_{k=1}^K \pi_k \mathcal{N}(\mathbf{x} | \boldsymbol\mu_k, \boldsymbol\Sigma_k)
$$
